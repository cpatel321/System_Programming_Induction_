4. Spectral Secret 

Given download file has an extension "npy" which is standard for exporting numpy array to binary format.

```python
# Importing all the necessary libraries and file which consist flag
import numpy as np
import matplotlib.pyplot as plt
data=np.load("spectral_secret.npy")
# shape of array
data.shape
```

Array consist of complex entries which gives a hint towards fourier transform.

```python
# inverse fourier transform of 2d data
dataf=np.fft.fft2(data)
# plotting the data
plt.imshow(np.abs(dataf),cmap='gray')
plt.show()
```

```python
# rotating the data by 180 degree
datafr=np.rot90(dataf,2)
# plotting the data

plt.imshow(np.abs(datafr),cmap='gray')
plt.show()
```
